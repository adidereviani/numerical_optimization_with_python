```python
import matplotlib.pyplot as plt
import numpy as np

# For n = 2 (2D simplex, which is a line segment)
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

x = np.linspace(0, 1, 100)
y = 1 - x
ax[0].plot(x, y, '-r')
ax[0].fill_between(x, 0, y, color='red', alpha=0.2)
ax[0].set_xlim(0, 1)
ax[0].set_ylim(0, 1)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('n=2 Probability Simplex')
ax[0].grid(True)

# For n = 3 (3D simplex, which is a triangle)
from mpl_toolkits.mplot3d import Axes3D
ax[1] = fig.add_subplot(122, projection='3d')
X = np.array([1, 0, 0])
Y = np.array([0, 1, 0])
Z = np.array([0, 0, 1])

# Vertices of the simplex triangle
vertices = np.array([[0, 0, 0], [1, 0, 0], [0, 1, 0], [0, 0, 1]])

# Triangles (indices of the vertices above)
triangles = [[0, 1, 2], [0, 1, 3], [0, 2, 3], [1, 2, 3]]

# Plotting the triangle representing the 3D simplex
ax[1].plot_trisurf(vertices[:, 0], vertices[:, 1], vertices[:, 2], triangles=triangles, color='blue', alpha=0.6, edgecolor='black')
ax[1].set_xlim(0, 1)
ax[1].set_ylim(0, 1)
ax[1].set_zlim(0, 1)
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].set_zlabel('z')
ax[1].set_title('n=3 Probability Simplex')
ax[1].view_init(elev=30, azim=30)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Define the radius
radius = 2

# For n = 2 (2D quarter circle)
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Define the circle
theta = np.linspace(0, np.pi/2, 100)  # From 0 to pi/2 to get the first quadrant
x = radius * np.cos(theta)
y = radius * np.sin(theta)

# Plot the quarter circle
ax[0].plot(x, y, '-b')
ax[0].fill_between(x, 0, y, color='blue', alpha=0.2)
ax[0].set_xlim(0, radius)
ax[0].set_ylim(0, radius)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('n=2 Quarter-circle (Non-negative entries)')
ax[0].grid(True)
ax[0].set_aspect('equal', adjustable='box')

# For n = 3 (3D octant of a sphere)
ax[1] = fig.add_subplot(122, projection='3d')
phi = np.linspace(0, np.pi/2, 50)  # Azimuth from 0 to pi/2
theta = np.linspace(0, np.pi/2, 50)  # Inclination from 0 to pi/2
phi, theta = np.meshgrid(phi, theta)

# Spherical to Cartesian conversion
x = radius * np.sin(theta) * np.cos(phi)
y = radius * np.sin(theta) * np.sin(phi)
z = radius * np.cos(theta)

# Plot the octant of the sphere
ax[1].plot_surface(x, y, z, color='green', alpha=0.6)
ax[1].set_xlim(0, radius)
ax[1].set_ylim(0, radius)
ax[1].set_zlim(0, radius)
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].set_zlabel('z')
ax[1].set_title('n=3 Sphere Octant (Non-negative entries)')
ax[1].view_init(elev=20, azim=45)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create a figure for plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the limits for x, y, z based on the linear constraints for the unit ball in the infinity norm
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
z = np.linspace(-1, 1, 100)

# Create meshgrids for X, Y, and Z planes
X, Y = np.meshgrid(x, y)
Z = np.ones_like(X)  # Upper and lower bounds for the cube at Z = 1 and Z = -1

# Plot each face of the cube
ax.plot_surface(X, Y, Z, color='b', alpha=0.5)    # Top face
ax.plot_surface(X, Y, -Z, color='b', alpha=0.5)   # Bottom face

X, Z = np.meshgrid(x, z)
Y = np.ones_like(X)  # Side faces at Y = 1 and Y = -1
ax.plot_surface(X, Y, Z, color='r', alpha=0.5)    # Front face
ax.plot_surface(X, -Y, Z, color='r', alpha=0.5)   # Back face

Y, Z = np.meshgrid(y, z)
X = np.ones_like(Y)  # Side faces at X = 1 and X = -1
ax.plot_surface(X, Y, Z, color='g', alpha=0.5)    # Right face
ax.plot_surface(-X, Y, Z, color='g', alpha=0.5)   # Left face

# Set plot limits and labels
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Unit Ball in R^3 for the Infinity Norm')

# Show the plot
plt.show()
```
